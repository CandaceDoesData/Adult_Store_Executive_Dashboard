```python
import pandas as pd
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
data = pd.read_csv('/Users/candacestingley/Documents/Portfolio_Projects/ASED_Data/raw_csv/Filtered_Super_Report.csv')
data
```

```python
data = data.dropna()
```

```python
data.info()
```

```python
df = pd.DataFrame(data)
```

```python
grouped = df.groupby('Department')
```

```python
department_dataframes = {}
```

```python
for department, group in grouped:
    department_dataframes[department] = group
```

```python
adorn_df = department_dataframes['ADORNMENTS']
app_df = department_dataframes['APPAREL']
app_pi_df = department_dataframes['APPAREL PIRATES']
bod_well_df = department_dataframes['BODY AND WELLNESS']
books_df = department_dataframes['BOOKS']
brand_df = department_dataframes['BRANDED PROMO PRODUCTS']
disp_df = department_dataframes['DISPLAY INVENTORY']
fant_wear_df = department_dataframes['FANTASY WEAR']
fet_df = department_dataframes['FETISH']
furn_df = department_dataframes['FURNITURE']
gifts_df = department_dataframes['GIFTS']
hose_df = department_dataframes['HOSIERY']
jewl_df = department_dataframes['JEWELRY']
lube_loti_df = department_dataframes['LUBES AND LOTIONS']
mens_df = department_dataframes['MENS PRODUCTS']
nonv_rub_df = department_dataframes['NONVIBES AND RUBBER']
res_misc_df = department_dataframes['RESORT AND MISCELLANEOUS']
vibes_df = department_dataframes['VIBES']
```

```python
adorn_plus = adorn_df[adorn_df['Qty Sold'] >= 0]
adorn_plus.info()
```

## Adornments

```python
adorn_bask_plus = (adorn_plus.groupby(['Transaction','Description'])['Qty Sold']
              .sum().unstack().reset_index().fillna(0).set_index('Transaction'))
adorn_bask_plus
```

```python
def encode_units(x):
    if x <= 0:
       return 0
    if x >= 1:
        return 1

adorn_bask_enc_plus = adorn_bask_plus.applymap(encode_units)
adorn_bask_enc_plus
```

```python
adorn_bask_filt_plus = adorn_bask_enc_plus[(adorn_bask_enc_plus > 0).sum(axis = 1) >= 2]
adorn_bask_filt_plus
```

```python
adorn_freq_item_plus = apriori(adorn_bask_filt_plus, min_support = 0.008,
                                use_colnames = True).sort_values('support', ascending = False).reset_index(drop = True)

adorn_freq_item_plus['length'] = adorn_freq_item_plus['itemsets'].apply(lambda x: len(x))

adorn_freq_item_plus
```

```python
association_rules(adorn_freq_item_plus, metric = 'lift',
                 min_threshold = 1).sort_values('lift', ascending = False).reset_index(drop = True)
```

```python
def encode_units(x):
    if x <= 0:
       return 0
    if x >= 1:
        return 1
toys_bask_enc_plus = toys_bask_plus.applymap(encode_units)
toys_bask_enc_plus
```

```python
toys_bask_filt_plus = toys_bask_enc_plus[(toys_bask_enc_plus > 0).sum(axis = 1) >= 2]
toys_bask_filt_plus
```
