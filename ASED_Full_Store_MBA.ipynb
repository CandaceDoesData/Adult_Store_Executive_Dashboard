```python
import pandas as pd
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
    data = pd.read_csv('/Users/candacestingley/Documents/Portfolio_Projects/ASED_Data/raw_csv/Filtered_Super_Report.csv')
data
```

```python
data = data.dropna()
```

```python
data.info()
```

```python
data_plus = data[data['Qty Sold'] >= 0]
data_plus.info()
```

```python
basket_plus = (data_plus.groupby(['Transaction','Description'])['Qty Sold']
              .sum().unstack().reset_index().fillna(0).set_index('Transaction'))
basket_plus
```

```python
def encode_units(x):
    if x <= 0:
       return 0
    if x >= 1:
        return 1

basket_encode_plus = basket_plus.applymap(encode_units)
basket_encode_plus
```

```python
basket_filter_plus = basket_encode_plus[(basket_encode_plus > 0).sum(axis = 1) >= 2]
basket_filter_plus
```

```python
frequent_itemsets_plus = apriori(basket_filter_plus, min_support = 0.0025,
                                use_colnames = True).sort_values('support', ascending = False).reset_index(drop = True)

frequent_itemsets_plus['length'] = frequent_itemsets_plus['itemsets'].apply(lambda x: len(x))

frequent_itemsets_plus
```

```python
association_rules(frequent_itemsets_plus, metric = 'lift',
                 min_threshold = 1).sort_values('lift', ascending = False).reset_index(drop = True)
```
